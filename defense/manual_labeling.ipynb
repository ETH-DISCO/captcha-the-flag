manually determine ground truth for segmentation task (there is no ground truth in the dataset, we can't use the models we want to benchmark to generate it)

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import random

# take a random file
root = Path.cwd() / "full captcha dataset"
files = [f for f in root.rglob("*.jpg")]
random_file = files[random.randint(0, len(files))]
cls = random_file.parts[-2]
print("class: ", cls)
print("path: ", random_file)

# load the image
img = plt.imread(random_file)
height, width, _ = img.shape

cell_height = height // 4
cell_width = width // 4

# plot the image
plt.figure(figsize=(10, 10))
for i in range(1, 4):
    line_width = 4
    plt.axhline(cell_height * i, color='white', linewidth=line_width)
    plt.axvline(cell_width * i, color='white', linewidth=line_width)
plt.imshow(img)
```
