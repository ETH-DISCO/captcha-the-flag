manually determine ground truth for segmentation task (there is no ground truth in the dataset, we can't use the models we want to benchmark to generate it)

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import random


root = Path.cwd() / "full captcha dataset"
files = [f for f in root.rglob("*.jpg")]
random_file = files[random.randint(0, len(files))]
cls = random_file.parts[-2]
print("class: ", cls)
print("path: ", str(random_file).replace(str(root), ""))

img = plt.imread(random_file)
height, width, _ = img.shape
cell_height = height // 4
cell_width = width // 4

plt.figure(figsize=(10, 10))
for i in range(1, 5):
    line_width = 4
    plt.axhline(cell_height * i, color='white', linewidth=line_width)
    plt.axvline(cell_width * i, color='white', linewidth=line_width)

    plt.text(cell_width * (i - 0.5), cell_height * 0.5, f"{i-1}".zfill(2), color='white', fontsize=20)
    plt.text(cell_width * (i - 0.5), cell_height * 1.5, f"{i+3}".zfill(2), color='white', fontsize=20)
    plt.text(cell_width * (i - 0.5), cell_height * 2.5, f"{i+7}".zfill(2), color='white', fontsize=20)
    plt.text(cell_width * (i - 0.5), cell_height * 3.5, f"{i+11}".zfill(2), color='white', fontsize=20)

plt.axis('off')
plt.imshow(img)
```
